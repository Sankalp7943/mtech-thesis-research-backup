## Handwritten Character Recognition 

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
tf.__version__
```

```python
#import datasets

mnist= tf.keras.datasets.mnist #28x28 images
(x_train,y_train),(x_test,y_test)=mnist.load_data()
#print(x_train[0])
plt.imshow(x_train[0])
```

```python
#normalizing the pixel values

x_train=tf.keras.utils.normalize(x_train,axis=1)
x_test=tf.keras.utils.normalize(x_test,axis=1)
plt.imshow(x_train[0])
x_train.shape
```

```python
#creating DL model
model=tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten())#numpy could be used to flatten the 2d array
model.add(tf.keras.layers.Dense(128,activation=tf.nn.relu))#neurons,activation func(rectified linear)
model.add(tf.keras.layers.Dense(128,activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(10,activation=tf.nn.softmax))#output layer (neurons=no of classes,softmax layer for probability distribution)
```

```python

model.build((None, 28, 28))
```

### Using mliiitl class to visualize model training against different optimisation techniques 
#### (In v4.0 this order is necessary, can be resolved later)
##### Order in v4
x_train, y_train, x_test, y_test, model, loss, epoch, batch_size

```python
# v5 fix ordering dependencies
mliiitl_object = Mliiitl(x_train, y_train, x_test, y_test, model, "sparse_categorical_crossentropy", 20, 64)
```

#### Hybrid class object (Mliiitl object, kwargs(optmisers = [rank, epochs])) 
##### order specific, rank starts from 1

```python
#creating DL model
model=tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten())#numpy could be used to flatten the 2d array
model.add(tf.keras.layers.Dense(128,activation=tf.nn.relu))#neurons,activation func(rectified linear)
model.add(tf.keras.layers.Dense(128,activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(10,activation=tf.nn.softmax))#output layer (neurons=no of classes,softmax layer for probability distribution)
```

```python
model.build((None, 28, 28))
```

```python
mliiitl_object_hybrid = Mliiitl(x_train, y_train, x_test, y_test, model, "sparse_categorical_crossentropy", 20, 64)
```

```python
hybrid_model_object = Hybrid(mliiitl_object_hybrid, adagrad=[1, 10], SGD = [2,10])
```

```python
ans = hybrid_model_object.run()
```

```python
# ans["loss"]
```

```python
plt.plot(ans['loss'])
plt.plot(ans["val_loss"])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.figure(figsize = (15,10))
plt.show()
```

```python
plt.plot(ans['acc'])
plt.plot(ans["val_acc"])
plt.title('model acc')
plt.ylabel('acc')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.figure(figsize = (15,10))
plt.show()
```
