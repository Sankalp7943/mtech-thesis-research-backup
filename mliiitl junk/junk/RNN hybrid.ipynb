```python
from mliiitl.mliiitl import Mliiitl
```

```python
from mliiitl.mliiitl import Hybrid #change package entry point to remove unneccessary long import
```

```python
from keras.models import Sequential
from keras.layers import Embedding, SimpleRNN
```

```python
from keras.datasets import imdb
from keras.preprocessing import sequence

max_features = 10000  # number of words to consider as features
maxlen = 500  # cut texts after this number of words (among top max_features most common words)
batch_size = 32

print('Loading data...')
(input_train, y_train), (input_test, y_test) = imdb.load_data(num_words=max_features)
print(len(input_train), 'train sequences')
print(len(input_test), 'test sequences')

print('Pad sequences (samples x time)')
input_train = sequence.pad_sequences(input_train, maxlen=maxlen)
input_test = sequence.pad_sequences(input_test, maxlen=maxlen)
print('input_train shape:', input_train.shape)
print('input_test shape:', input_test.shape)
```

```python
from keras.layers import Dense

model1 = Sequential()
model1.add(Embedding(max_features, 128))
model1.add(SimpleRNN(128))
model1.add(Dense(1, activation='sigmoid'))

# model1.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])
# history1 = model1.fit(input_train, y_train,
#                     epochs=20,
#                     batch_size=128,
#                     validation_split=0.2)
```

```python
mliiitl_object = Mliiitl(input_train, y_train, input_test, y_test, model1, "binary_crossentropy", 20, 64)
```

```python
mliiitl_object.test_performance(plots=True, save=False, splice=False, factor=1)
```

```python
model1 = Sequential()
model1.add(Embedding(max_features, 128))
model1.add(SimpleRNN(128))
model1.add(Dense(1, activation='sigmoid'))
```

```python
mliiitl_object_hybrid = Mliiitl(input_train, y_train, input_test, y_test, model1, "binary_crossentropy", 20, 64)
```

```python
hybrid_model_object = Hybrid(mliiitl_object_hybrid, adamax=[1, 2], adadelta = [2,8])
```

```python
ans = hybrid_model_object.run()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(ans['loss'])
plt.plot(ans["val_loss"])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.figure(figsize = (15,10))
plt.show()
```

```python
plt.plot(ans['acc'])
plt.plot(ans["val_acc"])
plt.title('model acc')
plt.ylabel('acc')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.figure(figsize = (15,10))
plt.show()
```
